```python
# ── Deterministic sampling: top-K users by train interaction count ────────────
TARGET_USERS           = 1000   # set to 1760 to use the full dataset
MIN_TRAIN_INTERACTIONS = 20     # users below this are dropped (same floor as ML-100K)
SAMPLED_DATA_DIR       = "hm_sampled"
```

```python

import os

def sample_hm_dataset_deterministic(train_df, test_df,
                                    target_users=1000,
                                    min_train_interactions=20,
                                    user_col="customer_id",
                                    item_col="product_code",
                                    target_col="bought"):
    """
    Deterministic user sampling — no randomness, fully reproducible.

    Steps:
      1. Drop users with fewer than min_train_interactions (hard floor).
      2. Rank remaining users by train interaction count, descending.
      3. Take the top target_users users (ties broken by user id, lexicographic).
      4. Induce item set from sampled users' train interactions only.
      5. Filter test to (sampled user, retained item) pairs.
      6. Re-index to contiguous 0-based integers.
    """

    # Step 1: enforce minimum interaction floor
    user_counts = (
        train_df.groupby(user_col)[item_col]
        .count().rename("n_interactions").reset_index()
    )
    user_counts = user_counts[user_counts["n_interactions"] >= min_train_interactions]

    # Step 2-3: rank by interaction count desc, break ties by user_id asc
    user_counts = user_counts.sort_values(
        ["n_interactions", user_col],
        ascending=[False, True]
    )
    sampled_users = user_counts.head(target_users)[user_col].values

    # Step 4: induce item set from sampled users' train rows only
    train_sub = train_df[train_df[user_col].isin(sampled_users)].copy()
    retained_items = train_sub[item_col].unique()

    # Step 5: filter test — only (sampled user, retained item) pairs
    test_sub = test_df[
        test_df[user_col].isin(sampled_users) &
        test_df[item_col].isin(retained_items)
    ].copy()

    # Step 6: re-index to 0-based contiguous integers
    # Sort by interaction count desc to make user 0 = most active (nice for debugging)
    ordered_users = (
        train_sub.groupby(user_col)[item_col].count()
        .sort_values(ascending=False).index
    )
    user2idx = {u: i for i, u in enumerate(ordered_users)}
    item2idx = {a: i for i, a in enumerate(retained_items)}

    for df_ in [train_sub, test_sub]:
        df_[user_col] = df_[user_col].map(user2idx)
        df_[item_col] = df_[item_col].map(item2idx)
        df_.dropna(subset=[user_col, item_col], inplace=True)
        df_[[user_col, item_col]] = df_[[user_col, item_col]].astype(int)

    train_sub.reset_index(drop=True, inplace=True)
    test_sub.reset_index(drop=True, inplace=True)

    meta = {
        "n_users": train_sub[user_col].nunique(),
        "n_items": train_sub[item_col].nunique(),
        "n_train": len(train_sub),
        "n_test":  len(test_sub),
        "n_users_with_test": test_sub[user_col].nunique(),
    }
    return train_sub, test_sub, meta

```

```python
# ── Save-or-load (no seed in cache tag — sampling is deterministic) ───────────
cache_tag  = f"u{TARGET_USERS}_m{MIN_TRAIN_INTERACTIONS}_det"
train_path = os.path.join(SAMPLED_DATA_DIR, f"train_{cache_tag}.csv")
val_path   = os.path.join(SAMPLED_DATA_DIR, f"val_{cache_tag}.csv")
test_path  = os.path.join(SAMPLED_DATA_DIR, f"test_{cache_tag}.csv")
meta_path  = os.path.join(SAMPLED_DATA_DIR, f"meta_{cache_tag}.csv")

if all(os.path.exists(p) for p in [train_path, val_path, test_path, meta_path]):
    print(f"Loading cached deterministic dataset from {SAMPLED_DATA_DIR}/")
    train_df = pd.read_csv(train_path)
    val_df   = pd.read_csv(val_path)
    test_df  = pd.read_csv(test_path)
    meta_df  = pd.read_csv(meta_path)
    N_USERS  = int(meta_df.loc[meta_df["key"] == "n_users",  "value"].iloc[0])
    N_ITEMS  = int(meta_df.loc[meta_df["key"] == "n_items",  "value"].iloc[0])
    print(f"  Loaded: {N_USERS} users | {N_ITEMS} items")
    print(f"  Train: {len(train_df)} | Val: {len(val_df)} | Test: {len(test_df)}")

else:
    print("Sampled dataset not found — running deterministic sampling pipeline...")

    train_df, test_df, meta = sample_hm_dataset_deterministic(
        train_df, test_df,
        target_users=TARGET_USERS,
        min_train_interactions=MIN_TRAIN_INTERACTIONS,
    )
    N_USERS = meta["n_users"]
    N_ITEMS = meta["n_items"]

    # Val split: last 20% of each user's train rows by item index (no randomness)
    # Sorting by (user, item) makes the split fully deterministic
    train_df_sorted = train_df.sort_values(["customer_id", "product_code"])
    val_rows = (
        train_df_sorted
        .groupby("customer_id", group_keys=False)
        .apply(lambda g: g.tail(max(1, round(len(g) * 0.2))))
    )
    val_df   = val_rows.reset_index(drop=True)
    train_df = train_df_sorted.drop(val_rows.index).reset_index(drop=True)

    # Save
    os.makedirs(SAMPLED_DATA_DIR, exist_ok=True)
    train_df.to_csv(train_path, index=False)
    val_df.to_csv(val_path,     index=False)
    test_df.to_csv(test_path,   index=False)
    pd.DataFrame([
        {"key": "n_users",             "value": N_USERS},
        {"key": "n_items",             "value": N_ITEMS},
        {"key": "n_train_after_val",   "value": len(train_df)},
        {"key": "n_val",               "value": len(val_df)},
        {"key": "n_test",              "value": meta["n_test"]},
        {"key": "n_users_with_test",   "value": meta["n_users_with_test"]},
        {"key": "target_users",        "value": TARGET_USERS},
        {"key": "min_train_interactions", "value": MIN_TRAIN_INTERACTIONS},
        {"key": "strategy",            "value": "deterministic_top_k"},
    ]).to_csv(meta_path, index=False)

    density = len(train_df) / (N_USERS * N_ITEMS) * 100
    print(f"  Users              : {N_USERS}")
    print(f"  Items              : {N_ITEMS}")
    print(f"  Train (after val)  : {len(train_df)} | Val: {len(val_df)} | Test: {len(test_df)}")
    print(f"  Users with test    : {meta['n_users_with_test']}")
    print(f"  Density            : {density:.4f}%")
```
