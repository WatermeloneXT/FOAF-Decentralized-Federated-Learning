```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the new data
rmse_df = pd.read_csv('ResultData/June30/order/rmse_table_order_random2out.csv')
time_df = pd.read_csv('ResultData/June30/order/time_table_order_random2out.csv')

plt.figure(figsize=(8,5))

# Plot for 5 orders
for i in range(5):
    rmse = rmse_df.iloc[i, 1:].astype(float).values
    time = time_df.iloc[i, 1:].astype(float).values.cumsum()
    min_len = min(len(time), len(rmse))
    plt.plot(time[:min_len], rmse[:min_len], label=f'Order {i+1}')

plt.xlabel('Accumulated Time')
plt.ylabel('RMSE')
plt.title('RMSE vs Accumulated Time (5 Orders)')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.savefig('ResultData/June30/order/order_cycle.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 4, figsize=(20, 10))  # 2 rows: directed, undirected

graph_types = ['random2out', 'random5out', 'scalefree', 'cycle']
titles = ['Random 2-Out', 'Random 5-Out', 'Scale-Free', 'Cycle']
versions = ['order_directed', 'order_undirected']  # row 0 = directed, row 1 = undirected

for row, version in enumerate(versions):
    for col, (graph_type, title) in enumerate(zip(graph_types, titles)):
        ax = axs[row, col]
        rmse_df = pd.read_csv(f'ResultData/June30/{version}/rmse_table_order_{graph_type}.csv')
        time_df = pd.read_csv(f'ResultData/June30/{version}/time_table_order_{graph_type}.csv')

        for i in range(5):
            rmse = rmse_df.iloc[i, 1:].astype(float).values
            time = time_df.iloc[i, 1:].astype(float).values.cumsum()
            min_len = min(len(time), len(rmse))
            ax.plot(time[:min_len], rmse[:min_len], label=f'Order {i+1}', linewidth=1.5)

        ax.set_xlabel('Accumulated Time')
        ax.set_ylabel('RMSE')
        if row == 0:
            ax.set_title(f'{title} (Directed)')
        else:
            ax.set_title(f'{title} (Undirected)')
        ax.grid(True)
        if i == 4:  # only add legend once per subplot
            ax.legend(fontsize='small')

plt.tight_layout()
plt.savefig('ResultData/June30/all_graphs_orders_directed_vs_undirected_2x4.png')
plt.show()
```
