```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
rmse_df = pd.read_csv('ResultData/June30/rmse_table_direc.csv')
time_df = pd.read_csv('ResultData/June30/time_table_direc.csv')

```

```python
# Extract RMSE and time values
rmse_directed = rmse_df.iloc[0, 1:].astype(float).values
rmse_undirected = rmse_df.iloc[1, 1:].astype(float).values
time_directed = time_df.iloc[0, 1:].astype(float).values.cumsum()
time_undirected = time_df.iloc[1, 1:].astype(float).values.cumsum()

# Align lengths
min_len = min(len(time_directed), len(rmse_directed), len(time_undirected), len(rmse_undirected))
time_directed = time_directed[:min_len]
rmse_directed = rmse_directed[:min_len]
time_undirected = time_undirected[:min_len]
rmse_undirected = rmse_undirected[:min_len]

```

```python
# Plot both in one figure
plt.figure(figsize=(8,5))
plt.plot(time_directed, rmse_directed, label='Directed')
plt.plot(time_undirected, rmse_undirected, label='Undirected')
plt.xlabel('Accumulated Time')
plt.ylabel('RMSE')
plt.title('RMSE vs Accumulated Time')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.savefig('ResultData/June30/directed_vs_undirected.png')
plt.show()

```
