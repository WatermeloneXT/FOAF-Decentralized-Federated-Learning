```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
rmse_df = pd.read_csv('ResultData/June30/rmse_table_direc.csv')
time_df = pd.read_csv('ResultData/June30/time_table_direc.csv')

```

```python
# Extract RMSE and time values
rmse_directed = rmse_df.iloc[0, 1:].astype(float).values
rmse_undirected = rmse_df.iloc[1, 1:].astype(float).values
time_directed = time_df.iloc[0, 1:].astype(float).values.cumsum()
time_undirected = time_df.iloc[1, 1:].astype(float).values.cumsum()

# Align lengths
min_len_directed = min(len(time_directed), len(rmse_directed))
min_len_undirected = min(len(time_undirected), len(rmse_undirected))

time_directed = time_directed[:min_len_directed]
rmse_directed = rmse_directed[:min_len_directed]
time_undirected = time_undirected[:min_len_undirected]
rmse_undirected = rmse_undirected[:min_len_undirected]
```

```python
# Plot Directed
plt.figure(figsize=(6,4))
plt.plot(time_directed, rmse_directed)
plt.xlabel('Accumulated Time')
plt.ylabel('RMSE')
plt.title('Directed')
plt.show()

# Plot Undirected
plt.figure(figsize=(6,4))
plt.plot(time_undirected, rmse_undirected)
plt.xlabel('Accumulated Time')
plt.ylabel('RMSE')
plt.title('Undirected')
plt.show()
```
