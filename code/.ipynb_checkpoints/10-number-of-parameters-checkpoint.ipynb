```python
def pprint(num):
    print(f"{num:,}")
```

```python
# ml100k 100: 100, 1160 
# ml100k: 943, 1628
# HM: 1760, 8618
n_users = 100
n_items = 1160
k = 7
l1 = 5
```

```python
n_data_points = n_users * n_items
pprint(n_data_points)
```

```python
mf = n_users * k + n_items * k
pprint(mf)
```

```python
dec_mf = (k + n_items * k)
pprint(dec_mf) # per user
pprint(n_users * dec_mf) # total
```

```python
gmf = (n_users * k + k * l1 * n_users) + (n_items * k + k * l1 * n_items)
pprint(gmf)
```

```python
dec_gmf = (k + k * l1 + n_items * k + n_items * k * l1)
pprint(dec_gmf) # per user
pprint(n_users * dec_gmf) # total
```

```python
gmf_shared = n_users * k + k * l1 + n_items * k + k * l1
pprint(gmf_shared)
```

```python
dec_gmf_shared = (k + k* l1 + n_items * k + k * l1)
pprint(dec_gmf_shared) # per user
pprint(n_users * dec_gmf_shared) # total
```
